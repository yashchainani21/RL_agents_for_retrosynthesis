```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = [
    "LM PKS products", "DORAnet bio products", "DORAnet chem products",
    "M1 PKS products", "DORAnet bio products", "DORAnet chem products",
    "M2 PKS products", "DORAnet bio products", "DORAnet chem products"
]

values = [
    31, 1866, 1806,
    1535, 260194, 160589,
    67192, 23129521, 10862072
]

# Bar positions with spacing between generations
positions = [0, 1, 2, 4, 5, 6, 8, 9, 10]

# Plot
plt.figure(figsize=(12, 6))
bars = plt.bar(positions, values, color='slategray', edgecolor='black')

# X-axis labels
plt.xticks(positions, labels, rotation=45, ha='right')

# Use log scale for y-axis
plt.yscale('log')

# Add value labels on top of bars (log-scaled spacing)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height * 1.15,
             f'{height:,}', ha='center', va='bottom', fontsize=8)

# Labels and title
plt.title('Number of Products per Generation (Log Scale)')
plt.ylabel('Number of Products (log₁₀ scale)')
plt.grid(axis='y', which='both', linestyle='--', alpha=0.5)

# Layout
plt.tight_layout()
plt.show()
```
