```python
import doranet.modules.enzymatic as enzymatic
import doranet.modules.synthetic as synthetic
import doranet.modules.post_processing as post_processing
```

```python
target_name = "test"
canon_target_smiles = "C=CC(O)=O"
num_generations = 1
job_name = f'{target_name}_CHEM{num_generations}_ONLY_RETRO'
chemistry_helpers = ("O", "O=O", "[H][H]", "O=C=O", "C=O", "[C-]#[O+]", "Br", "[Br][Br]", "CO", "[C-]#[O+]", "C=C", "O=S(O)O", "N", "O=S(=O)(O)O", "O=NO", "N#N", "O=[N+]([O-])O", "NO", "C#N", "S", "O=S=O", "N#CO")
```

```python
reverse_network = synthetic.generate_network(
            job_name = job_name,
            starters = {"C=CC(O)=O"},
            helpers = chemistry_helpers,
            gen = num_generations,
            direction = 'retro',
        )
```

```python
common_metabolites = ["C=CC=O","C=C=C(O)O"]

post_processing.pretreat_networks(networks = {reverse_network},
                                          total_generations = num_generations,
                                          starters = {x for x in common_metabolites},
                                          helpers = chemistry_helpers,
                                          job_name = job_name)
```

```python
post_processing.pathway_finder(starters = {x for x in common_metabolites},
                                       helpers = chemistry_helpers,
                                       target = canon_target_smiles,
                                       search_depth = num_generations,
                                       max_num_rxns = num_generations,
                                       min_rxn_atom_economy = 0,
                                       job_name = job_name)
```
